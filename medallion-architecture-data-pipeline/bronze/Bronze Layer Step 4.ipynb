```
from pyspark.sql.functions import trim, initcap, col

# Read the Delta table
df = spark.read.format("delta").load("dbfs:/FileStore/MedallionArchitecture/bronze/Customers")

# Transform the data
df_transformed = df.withColumn("CompanyName", initcap(trim(col("CompanyName")))).withColumn("ContactName", initcap(trim(col("ContactName")))).withColumn("City", initcap(trim(col("City")))).withColumn("Country", initcap(trim(col("Country"))))

# Check for NULL values in CategoryID
null_category_id_df = df_transformed.filter(col("CustomerID").isNull())

# Show rows where CategoryID is NULL
if null_category_id_df.count() > 0:
    print("Rows with NULL CategoryID found:")
    null_category_id_df.show()
else:
    print("No NULL values found in CategoryID.")

# Show the transformed DataFrame
display(df_transformed)
```

```
df_transformed.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save("dbfs:/FileStore/MedallionArchitecture/bronze/ArticleCategories")
```
